```python
import numpy as np
```

```python
N = 1000000
s = np.random.uniform(0,1,N)
```

```python
s.shape
eps = 0.01
```

```python
theta = 1.0
cnt = 0

for i in range(N):
    if s[i] <= theta:
        cnt += 1
        theta = theta*(1 - eps)
    else:
        theta = theta*(1 + eps)
        
print(cnt/N)
```
